```python
def eligible():
    gender = input("Your Gender: ")
    age = int(input("Your Age: "))
    if gender.lower() == "male":
        if age >= 21:
            print("eligible")
        else:
            print("not eligible")
    elif gender.lower() == "female":
        if age >= 18:        
            print("eligible")
        else:                
            print("not eligible")

eligible()
```
